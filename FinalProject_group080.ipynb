Felony and misdemeanor classifications vary across states. They differ based on the severity of the crime and punishment if convicted. California law uses criminal statutes which assign fixed punishments specific to each crime. Felonies are defined as more serious, violent crimes and can be punishable by incarceration in state prison, life imprisonment, and in the most serious cases, the death penalty. Misdemeanors include less serious, non-violent offenses that can be punishable by fines, imprisonment in county jail, or infractions.¹ Common examples of felonies under California Law include robbery, murder, grand theft, and rape.2 Common examples of misdemeanors include assault, battery, petty theft, and reckless driving.3 Crime can also be classified as either violent or property crime.

There are more than 10 million property crimes reported annually, while only about 1 million violent crimes5. Up next are motor vehicle theft (about 10%), aggravated assault (about 7 %), and robbery (violent theft, directly on a person or place, about 3% of all reported crimes).

# Hypothesis


*We hypothesize that each subtype of violent crime will be more prevalent in more densely populated counties, while each subtype of non-violent crimes will be more prevalent in less densely populated counties. We assert this claim because the more densely populated counties will result in higher tensions due to close proximity, and less densely populated counties are more spread out so there will be more non-contact crimes. The reasoning is that with a higher population density, there are more interactions and a higher probability of negative interactions and the inverse is likely in lower population density areas. From this, it can be reasoned that most interactions, especially negative ones between humans, and property would be more likely where there is a higher ratio of property to humans.*

```python
##Normalizing Stats by type of density (High, Medium, Low)
#These numbers were manually added up from the tables(Divide by population of each density type, multiply by 100000)

def LowNorm(value):
    value = value / 1062912 * 100000
    return value

def MedNorm(value):
    value = value / 7307350 * 100000
    return value

def HighNorm(value):
    value = value / 31293866 * 100000
    return value
```

```python
#Comparing proprotions of sub-categories of crime for 2019 and 2014
Stats2019CADrop = Stats2019CA.drop(columns=['Violent Crime', 'Property crime'])
Stats2017CADrop = Stats2017CA.drop(columns=['Violent Crime', 'Property crime'])
Stats2014CADrop = Stats2014CA.drop(columns=['Violent Crime', 'Property crime'])

#Concatonated Dataframe of the Crime Statistics
concat_2014_2019 = pd.concat({'2014': Stats2014CADrop, '2017':Stats2017CADrop, '2019': Stats2019CADrop})

#Concatonated Dataframe Clean Up
df1 = concat_2014_2019.reset_index ()
df2 = df1.drop(columns=['level_1'])
df3 = df2.rename(columns={'level_0': 'Year'})

#Finding the Averages of the Crimes Rates over the Decade
concat_2014_2019Clean = df3.groupby('Year').mean()
concat_2014_2019Clean
```

```python
#Ploting the different years to show difference in crime category over time
concat_2014_2019Clean.plot(kind='bar', stacked=True)
plt.title('Difference in category of crime: 2014, 2017, 2019')
plt.xlabel('Year')
plt.ylabel('Crimes')
```

Just like we saw before in the linplot, the rate of crime has been going down in general, but some of the Violent Crime rates are going up such as Aggravated Assault (Red). Larceny-Theft seems to be one of the most prevalent types, going up in reccent years as well.
